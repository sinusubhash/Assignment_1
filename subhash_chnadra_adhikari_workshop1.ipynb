Problem - 1: Array Creation:

```python
import numpy as np

# 1. Initialize an empty array with size 2x2
empty_array = np.empty((2, 2))
print("1. Empty Array (2x2):")
print(empty_array, "\n")

# 2. Initialize an all-one array with size 4x2
ones_array = np.ones((4, 2))
print("2. All-Ones Array (4x2):")
print(ones_array, "\n")

# 3. Return a new array of given shape and type, filled with fill value
fill_value_array = np.full((3, 3), 7)  # Example: 3x3 array filled with 7
print("3. Array Filled with Value 7 (3x3):")
print(fill_value_array, "\n")

# 4. Return a new array of zeros with the same shape and type as a given array
reference_array = np.array([[5, 6], [7, 8]])
zeros_like_array = np.zeros_like(reference_array)
print("4. Zeros Array with Same Shape as Reference Array:")
print(zeros_like_array, "\n")

# 5. Return a new array of ones with the same shape and type as a given array
ones_like_array = np.ones_like(reference_array)
print("5. Ones Array with Same Shape as Reference Array:")
print(ones_like_array, "\n")

# 6. Convert a list to a NumPy array
new_list = [1, 2, 3, 4]
converted_array = np.array(new_list)
print("6. Converted List to NumPy Array:")
print(converted_array)
```

Problem - 2: Array Manipulation: Numerical Ranges and Array indexing:

```python
import numpy as np

# 1. Create an array with values ranging from 10 to 49
array_range = np.arange(10, 50)
print("1. Array with values ranging from 10 to 49:")
print(array_range, "\n")

# 2. Create a 3x3 matrix with values ranging from 0 to 8
matrix_3x3 = np.arange(9).reshape(3, 3)
print("2. 3x3 Matrix with values ranging from 0 to 8:")
print(matrix_3x3, "\n")

# 3. Create a 3x3 identity matrix
identity_matrix = np.eye(3)
print("3. 3x3 Identity Matrix:")
print(identity_matrix, "\n")

# 4. Create a random array of size 30 and find the mean of the array
random_array = np.random.random(30)
mean_value = random_array.mean()
print("4. Random Array of size 30 and its Mean:")
print("Random Array:", random_array)
print("Mean:", mean_value, "\n")

# 5. Create a 10x10 array with random values and find the min and max values
random_matrix = np.random.random((10, 10))
min_value = random_matrix.min()
max_value = random_matrix.max()
print("5. 10x10 Random Array with Min and Max Values:")
print("Minimum Value:", min_value)
print("Maximum Value:", max_value, "\n")

# 6. Create a zero array of size 10 and replace 5th element with 1
zero_array = np.zeros(10)
zero_array[4] = 1  # Replace 5th element (index 4 in 0-based indexing)
print("6. Zero Array with 5th Element Replaced by 1:")
print(zero_array, "\n")

# 7. Reverse an array arr = [1, 2, 0, 0, 4, 0]
arr = [1, 2, 0, 0, 4, 0]
reversed_array = np.array(arr)[::-1]
print("7. Reversed Array:")
print(reversed_array, "\n")

# 8. Create a 2D array with 1 on border and 0 inside
border_array = np.ones((5, 5))
border_array[1:-1, 1:-1] = 0
print("8. 2D Array with 1 on Border and 0 Inside:")
print(border_array, "\n")

# 9. Create an 8x8 matrix and fill it with a checkerboard pattern
checkerboard = np.zeros((8, 8), dtype=int)
checkerboard[1::2, ::2] = 1  # Fill alternate cells in odd rows
checkerboard[::2, 1::2] = 1  # Fill alternate cells in even rows
print("9. 8x8 Checkerboard Pattern Matrix:")
print(checkerboard)
```

Problem - 3: Array Operations:

```python
import numpy as np

# Initialize arrays
x = np.array([[1, 2], [3, 5]])
y = np.array([[5, 6], [7, 8]])
v = np.array([9, 10])
w = np.array([11, 12])

# 1. Add the two arrays
addition_result = x + y
print("1. Addition of x and y:")
print(addition_result, "\n")

# 2. Subtract the two arrays
subtraction_result = x - y
print("2. Subtraction of x and y:")
print(subtraction_result, "\n")

# 3. Multiply the array `x` with an integer (e.g., 2)
scalar_multiplication = x * 2
print("3. Multiplication of x with 2:")
print(scalar_multiplication, "\n")

# 4. Find the square of each element of the array `x`
squared_elements = np.square(x)
print("4. Square of each element in x:")
print(squared_elements, "\n")

# 5. Find the dot product
dot_product_vw = np.dot(v, w)
dot_product_xv = np.dot(x, v)
dot_product_xy = np.dot(x, y)
print("5. Dot Products:")
print(f"Dot product of v and w: {dot_product_vw}")
print(f"Dot product of x and v: {dot_product_xv}")
print(f"Dot product of x and y:\n{dot_product_xy}\n")

# 6. Concatenate x and y along rows, and v and w along columns
concat_xy_row = np.concatenate((x, y), axis=0)  # Along rows
concat_vw_col = np.vstack((v, w))  # Along columns
print("6. Concatenation Results:")
print("Concatenation of x and y along rows:")
print(concat_xy_row)
print("Concatenation of v and w along columns:")
print(concat_vw_col, "\n")
```

Problem - 4: Matrix Operations:

```python
import numpy as np

# Initialize matrices A and B
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# 1. Prove A.A⁻¹ = I
A_inv = np.linalg.inv(A)
identity_matrix = np.dot(A, A_inv)
print("1. Prove A.A⁻¹ = I:")
print("A * A⁻¹:")
print(identity_matrix)
print("Is A * A⁻¹ an Identity Matrix? ", np.allclose(identity_matrix, np.eye(2)), "\n")

# 2. Prove AB ≠ BA
AB = np.dot(A, B)
BA = np.dot(B, A)
print("2. Prove AB ≠ BA:")
print("AB:")
print(AB)
print("BA:")
print(BA)
print("Is AB equal to BA? ", np.array_equal(AB, BA), "\n")

# 3. Prove (AB)^T = B^T * A^T
AB_T = np.transpose(AB)
BT_AT = np.dot(np.transpose(B), np.transpose(A))
print("3. Prove (AB)^T = B^T * A^T:")
print("(AB)^T:")
print(AB_T)
print("B^T * A^T:")
print(BT_AT)
print("Is (AB)^T equal to B^T * A^T? ", np.array_equal(AB_T, BT_AT), "\n")

# System of Linear Equations
# 2x − 3y + z = −1
# x − y + 2z = −3
# 3x + y − z = 9

# Represent the equations in matrix form AX = B
A_eq = np.array([[2, -3, 1], [1, -1, 2], [3, 1, -1]])  # Coefficients of variables
B_eq = np.array([-1, -3, 9])  # Constants on the right-hand side

# Solve for X using inverse methods
A_inv_eq = np.linalg.inv(A_eq)
X_solution = np.dot(A_inv_eq, B_eq)

print("4. Solve System of Linear Equations Using Inverse Methods:")
print("Solution [x, y, z]:")
print(X_solution, "\n")

# Solve using np.linalg.solve (preferred method)
X_solution_solve = np.linalg.solve(A_eq, B_eq)
print("Solution Using np.linalg.solve:")
print(X_solution_solve)
```
